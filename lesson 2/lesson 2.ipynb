```python
bookings_churn_rate_df = bookings\
.groupby(['has_kids','is_canceled'])[['is_canceled']]\
.count()\
.rename(columns={'is_canceled':'number_of_bookings'})\
.unstack()
bookings_churn_rate_df
```

```python
def find_churn_rate(row):
    cancelled_bookings = row['number_of_bookings'][1]
    all_bookings = row.sum()
    row['churn_rate'] = round(100 * cancelled_bookings / all_bookings, 2)
    return row
```

```python
bookings_churn_rate_df = bookings_churn_rate_df.apply(find_churn_rate,axis=1)
```
