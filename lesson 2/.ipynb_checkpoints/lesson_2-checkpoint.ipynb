##### 1.3.Определить суммарные продажи курсов с группировкой данных по городам.

````python
income_by_city = df.groupby(['title','city'], as_index=False)\
            .agg({'benefit':'sum'})\
            .rename(columns={'title':'title','benefit':'sum_income'})\
            .sort_values(by='sum_income', ascending = False)
income_by_city.head()
````

##### 1.4.Определить суммарные продажи и количество завершённых курсов с группировкой по их названию.

````python
courses_stats = df.query('status == "Завершен"')\
            .groupby(['title'], as_index=False)\
            .agg({'benefit':'sum', 'number':'count'})\
            .rename(columns={'title':'title','benefit':'sum_income','number':'number_of_positive_agreements'})\
            .sort_values(by='sum_income', ascending = False)
courses_stats.head()
````

##### 2.10.Напишите функцию `temp_to_celcius`, которая получает на вход колонку с температурой в °F и возвращает значения, переведённые в градусы Цельсия.
Формула: 
$$C = \frac{(F-32) \cdot 5.0}{9.0}$$

````python
taxi = taxi.rename(columns={'temp':'temp_F'})
taxi.head()
````

````python
def temp_to_celcius(temp_F):
    temp_C = round((temp_F - 32)*5/9,2)
    return temp_C
````

````python
taxi['temp_C'] = taxi.temp_F.apply(temp_to_celcius)
taxi.head()
````

##### 3.3.Пользователи из каких стран совершили наибольшее число успешных бронирований? Указать топ-5.

````python
bookings\
.query('is_canceled == 0')\
.groupby('country', as_index=False)\
.agg({'is_canceled':'count'})\
.sort_values('is_canceled',ascending=False)\
.rename(columns = {'is_canceled':'successful_bookings'})\
.head()
````

##### 3.4.На сколько ночей (stays_total_nights)  в среднем бронируют отели типа City Hotel и Resort Hotel? Записать полученные значения в пропуски с точностью до 2 знаков после запятой.

````python
bookings\
.query('(hotel == "City Hotel") | (hotel == "Resort Hotel")')\
.groupby('hotel')\
.agg({'stays_total_nights':'mean'})\
.round({'stays_total_nights':2,})
````

##### 3.6.Теперь проанализируйте даты запланированного прибытия (arrival_date_year). <br> <br> На какой месяц чаще всего оформляли бронь в 2016 году? <br> <br> Изменился ли самый популярный месяц в 2017 году? <br> <br> В 2015?

````python
# top_arrive_month_2016 = bookings\
# .query('arrival_date_year == 2016')\
# .groupby('arrival_date_month')\
# .agg({'hotel':'count'})\
# .rename(columns={'hotel':'number_of_bookings'})\
# .idxmax()\
# .values[0]

# top_arrive_month_2017  = bookings\
# .query('arrival_date_year == 2017')\
# .groupby('arrival_date_month')\
# .agg({'hotel':'count'})\
# .rename(columns={'hotel':'number_of_bookings'})\
# .idxmax()\
# .values[0]

# if top_arrive_month_2016 == top_arrive_month_2017:
#     print(f'{top_arrive_month_2016} is one of the most popular months both in 2016 and 2017.')
# else:
#     print(f'''{top_arrive_month_2016} is the most popular month in 2016.
#     {top_arrive_month_2017} is the most popular month in 2017.''')
````

````python
popular_months_all_years = bookings\
                      .groupby(['arrival_date_year', 'arrival_date_month'], as_index = False)\
                      .agg({'hotel':'count'})\
                      .sort_values(by=['arrival_date_year', 'hotel'], ascending=[True, False])\
                      .rename(columns = {'hotel':'number_of_reservations'})\
                      .groupby('arrival_date_year')\
                      .head(1)

years_to_compare = [2015, 2016, 2017]

popular_months_selected_years = popular_months_all_years\
        .loc[popular_months_all_years.arrival_date_year.isin(years_to_compare)]

if (
        popular_months_selected_years\
        .arrival_date_month\
        .drop_duplicates()\
        .shape[0] == 1
    ):
    print(f"""Top month hasn't changed during {', '.join(list(map(str, years_to_compare)))}.  
          {popular_months_selected_years.arrival_date_month.drop_duplicates().values} is one of the most popular months.""")
else:
    popular_months_selected_years\
    .apply(lambda x: print(f'{x.arrival_date_month} is the most popular month in {x.arrival_date_year}.'), axis=1)
````

##### 3.7.Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов.

##### 3.10.Создайте переменную `has_kids`, которая принимает значение `True`, если клиент при бронировании указал хотя бы одного ребенка (total_kids), и False – в противном случае. Посчитайте отношение количества ушедших пользователей к общему количеству клиентов, выраженное в процентах (`churn rate`). Укажите, среди какой группы показатель выше.

````python
bookings['has_kids'] = bookings.total_kids != 0
bookings.head()
````

````python
canceled_bokings = bookings.is_canceled.sum()
all_bokings = bookings.shape[0]
total_churn_rate = round(canceled_bokings / all_bokings, 2)
print(f'total_churn_rate = {total_churn_rate}')
````

````python
churn_rate_df = pd.pivot_table(data = bookings,
               values='hotel',
               index='has_kids', 
               columns='is_canceled', 
               aggfunc='count')\
.apply(lambda x: round(100*x[1]/(x[0] + x[1]),2), axis=1).to_frame()\
.rename(columns = {0:'churn_rate'})

churn_rate_df.head()
````

````python
if churn_rate_df.churn_rate.idxmax():
    filler = 'with'
else:
    filler = 'without'
print(f'People {filler} children have the largest charn rate which is equal to {churn_rate_df.churn_rate.max()}%.')
````
